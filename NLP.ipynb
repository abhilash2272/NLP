**Removing Punctuations**

```python
import string
print(string.punctuation)

clean_text = text.translate(str.maketrans('', '', string.punctuation))
print(clean_text)

from nltk.tokenize import RegexpTokenizer
tokenizer = RegexpTokenizer(r'\w+')
clean_tokens = tokenizer.tokenize(clean_text)
print(clean_tokens)
```

**Stemming**

Stemming reduces words to their root form.

Process of removing prefix and suffix.

```python
from nltk.stem import PorterStemmer
porter_stemmer = PorterStemmer()
stemmed_tokens = [porter_stemmer.stem(token) for token in clean_tokens]
print(stemmed_tokens)
```

```python
# Snowball Stemmer : is an enhanced version of the Porter Stemmer, Supports Multiple Language
from nltk.stem import SnowballStemmer
snowball_stemmer = SnowballStemmer(language='english')
stemmed_tokens = [snowball_stemmer.stem(token) for token in clean_tokens]
print(stemmed_tokens)
```

```python
# Lancaster Stemmer : is known for being more aggresive and faster than other stemmers.
from nltk.stem import LancasterStemmer
lancaster_stemmer = LancasterStemmer()
stemmed_tokens = [lancaster_stemmer.stem(token) for token in clean_tokens]
print(stemmed_tokens)
```

**Lemmatization**

## **Lemmatization**

Lemmatization produces meaningful root words.

**Example**

better → good

running → run

Lemmatization is an important text pre-processing technique in Natural Language Processing (NLP) that reduces words to their base form known as a "lemma." For example, the lemma of "running" is "run" and "better" becomes "good."

Unlike stemming which simply removes prefixes or suffixes, it considers the word's meaning and part of speech (POS) and ensures that the base form is a valid word. This makes lemmatization more accurate as it avoids generating non-dictionary words.

**Used for:**

Improves accuracy

Reduced Data Redundancy

Better NLP Model Performance

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
lemmatized_tokens = [lemmatizer.lemmatize(token) for token in clean_tokens]
print(lemmatized_tokens)
print(lemmatizer.lemmatize("running",pos='v'))
```

## **Bag of Words (BoW)**

BoW represents text as word frequency vectors.

In Natural Language Processing (NLP) text data needs to be converted into numbers so that machine learning algorithms can understand it. One common method to do this is Bag of Words (BoW) model. It turns text like sentence, paragraph or document into a collection of words and counts how often each word appears but ignoring the order of the words. It does not consider the order of the words or their grammar but focuses on counting how often each word appears in the text.

This makes it useful for tasks like text classification, sentiment analysis and clustering.


**Step 1: Preprocessing the text**

Clean the text

 Convert to lowercase

 Remove special characters

```python
import re

dataset = []
text = [
    "I love coding!",
    "Machine learning is fun.",
    "Natural language processing is interesting$."
]
for sentence in text:
  sentence = sentence.lower()
  sentence = re.sub(r'[^\w\s]', '', sentence)
  dataset.append(sentence)

print(dataset)
```

**Step 2:Counting Word Frequencies**

```python
from collections import defaultdict
word_freq = defaultdict(int)
for sentence in dataset:
  words = sentence.split()
  for word in words:
    word_freq[word] +=1
print(word_freq)
```

**Step 3:Selecting Most Frequent Words**

```python
import heapq
freq_words = heapq.nlargest(3, word_freq, key=word_freq.get)
print(freq_words)
```

Step 4: Build Bag of Words Matrix

Now we create a matrix:

1-->Present

0-->absent

```python
x = []
for sentence in dataset:
  vector = []
  words = sentence.split()

  for word in freq_words:
    if word in words:
      vector.append(1)
    else:
      vector.append(0)
  x.append(vector)
print(x)
```

# **N-grams**

N-grams capture word sequences.

N-gram is a technique that splits text into sequences of N consecutive words to capture context and word order for machine learning models.

N-gram = group of N consecutive words

1 word → Unigram

2 words → Bigram

3 words → Trigram


#### **✅ Simple Example**

Sentence:

#### "I love coding"
####🔹 Unigrams (N = 1)
["I", "love", "coding"]
####🔹 Bigrams (N = 2)
["I love", "love coding"]
####🔹 Trigrams (N = 3)
["I love coding"]

```python
import re

dataset = []
text = [
    "I love coding!",
    "Machine learning is fun.",
    "Natural language processing is interesting$."
]
for sentence in text:
  sentence = sentence.lower()
  sentence = re.sub(r'[^\w\s]', '', sentence)
  dataset.append(sentence)

print(dataset)
```

```python
# create n-grams
def generate_ngrams(sentence,n):
  words = sentence.split()
  ngrams = []
  for i in range(len(words)-n+1):
    ngrams.append(" ".join(words[i:i+n]))
  return ngrams
```

```python
# Example :Bigrams(N=2)
for sentence in dataset:
  bigrams = generate_ngrams(sentence,2)
  print(bigrams)

# Example Trigrams(N=3)
for sentence in dataset:
  trigrams = generate_ngrams(sentence,3)
  print(trigrams)
```

```python
from collections import defaultdict
ngram2count = defaultdict(int)
for sentence in dataset:
  ngram = generate_ngrams(sentence,3)
  for n in ngram:
    ngram2count[n]+=1

print(ngram2count)
```

```python
# Convert into feature vector
import heapq
freq_ngrams = heapq.nlargest(3, ngram2count, key=ngram2count.get)
print(freq_ngrams)
```

```python
# Build Matrix
x = []
for sentence in dataset:
  vector = []
  ngrams = generate_ngrams(sentence,3)
  for ngram in freq_ngrams:
    if ngram in ngrams:
      vector.append(1)
    else:
      vector.append(0)
  x.append(vector)
print(x)
```

```python
# CBOW
from gensim.models import Word2Vec

sentences = [
    ['i', 'love', 'coding'],
    ['i', 'love', 'python'],
    ['python', 'is', 'powerful']
]
# Train CBOW model(sg=0)
cbow_model = Word2Vec(sentences,
                      vector_size=5,
                      window=2,
                      min_count=1,
                      sg=0
)

# Get Word Vector
print("Word Vector for 'coding':")
print(cbow_model.wv['coding'])

# Find similar words
similar_words = cbow_model.wv.most_similar('coding', topn=2)
print("Similar Words to 'coding':")
print(similar_words)
```

## **🎯 What is NER?**

👉 Named Entity Recognition (NER) is the process of identifying and classifying important words in a sentence into categories like:

Person 👤

Location 🌍

Organization 🏢

Date 📅

Money 💰

**Example**

Sentence:

"Elon Musk founded Tesla in California in 2003"

NER Output:

Elon Musk  → PERSON  
Tesla      → ORGANIZATION  
California → LOCATION  
2003       → DATE  


```python
sentence = "Elon Musk founded Tesla in California in 2003"
words = nltk.word_tokenize(sentence)
pos_tags = nltk.pos_tag(words)
print(pos_tags)

ner_tags = nltk.ne_chunk(pos_tags)
print(ner_tags)
```
